```python
from pathlib import Path
import pandas as pd
import json
import plotly.express as px
```

```python
# Retorna a pasta onde o notebook está a correr
ROOT = Path.cwd().resolve()
print(f"A raiz (no notebook) é: {ROOT}")
```

```python
# 1. Defina o caminho completo até o ARQUIVO
path = ROOT / 'Timeline.json'

with open(path, 'r') as f:
    file = json.load(f)
```

```python
file
```

```python
lista_pontos = []

for segmento in file.get('semanticSegments',[]):
    if 'timelinePath' in segmento:
        for info in segmento['timelinePath']:
            # O ponto vem como string: '-23.5387615°, -46.5278477°'
            # Vamos limpar os caracteres especiais e separar lat/lon
            coord_str = info['point'].replace('Â°', '').replace('°', '')
            lat, lon = map(float, coord_str.split(','))

            lista_pontos.append({
                'timestamp': pd.to_datetime(info['time']),
                'lat': lat,
                'lon': lon
            })

```

```python
df = pd.DataFrame(lista_pontos)
df
```

```python
# 3. Filtrar Setembro (opcional)
df = df[df['timestamp'].dt.month == 9]

if df.empty:
    print("Nenhum ponto encontrado. Verifique se o filtro de data está correto.")
else:
    # 4. Criar o Mapa Interativo
    fig = px.scatter_mapbox(
        df, 
        lat="lat", 
        lon="lon", 
        hover_data=["timestamp"],
        zoom=12, 
        height=600,
        title="Minha Viagem - Setembro 2025"
    )

    # Estilo do mapa (OpenStreetMap não precisa de chave de API)
    fig.update_layout(mapbox_style="open-street-map")
    fig.update_layout(margin={"r":0,"t":40,"l":0,"b":0})

    # 5. Salvar e Abrir
    nome_mapa = "mapa_viagem_setembro.html"
    fig.write_html(nome_mapa)
    print(f"Mapa gerado com sucesso! Procure o arquivo '{nome_mapa}' na sua pasta e abra no navegador.")
```
